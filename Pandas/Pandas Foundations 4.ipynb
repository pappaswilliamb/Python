# (2) Constructing Multiple Boolean Conditions

```python
movie = pd.read_csv('movie.csv', index_col = 'movie_title')
movie.head()
```

```python
criteria1 = movie['imdb_score'] > 8
criteria1
```

```python
criteria2 = movie['content_rating'] == 'PG-13'
criteria2
```

```python
criteria3 = (movie['imdb_score'] > 8) & (movie['content_rating'] == 'PG-13')
criteria3
```

```python
criteria4 = (movie['imdb_score'] > 8) | (movie['content_rating'] == 'PG-13')
criteria4
```

```python
criteria5 = (movie['imdb_score'].gt(8)) | (movie['content_rating'].eq('PG-13'))
criteria5
```

# (3) Filtering With Boolean Indexing

```python
criteria3.value_counts()
```

```python
movie[criteria3]
```

```python
movie.loc[criteria3]
```

```python
movie_filtered = movie[criteria3][['imdb_score', 'content_rating']]
movie_filtered
```

```python
movie_filtered = movie.loc[criteria3, ['imdb_score', 'content_rating']]
movie_filtered
```
