## (3) What are the ten most popular hashtags?

```python
hashtag_list = []
for t in data:
    for h in t['entities']['hashtags']:
        if h != []:
            hashtag_list.append(h['text'])
```

```python
pprint(hashtag_list)
```

```python
c = Counter(hashtag_list)
```

```python
c.most_common(10)
```

## (4) Who is the most frequently tweeting person about the topic?

```python
name_list = []
for t in data:
    if t != '':
        name_list.append(t['user']['name'])
```

```python
pprint(name_list)
```

```python
c = Counter(name_list)
```

```python
c.most_common(10)
```

# WordClouds

## (1) Visualization with WordCloud of 'frankenstein.txt'

```python
# Install 
!pip install wordcloud
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
infile = open('frankenstein.txt')
contents = infile.read()
infile.close()
```

```python
# Generating a WordCloud image
wordcloud = WordCloud(max_font_size = 80, collocations = False).generate(contents) # Setting lower max_font_size to 80
```

```python
# Display the generated image:
%matplotlib inline
plt.figure(figsize = (15, 10))
plt.imshow(wordcloud)
plt.axis("off")
plt.savefig('WordCloud1.pdf')
plt.show()
```

## (2) Create a WordCloud from Tweets

```python
import json
from wordcloud import WordCloud
import matplotlib.pyplot as plt
%matplotlib inline

#(1) Open the JSON file and create a string variable, content
infile = open('tweet_stream_covid_200.json')
data = json.load(infile)
infile.close()

contents = ''

for t in data:
    contents += t['text'] + '\n'
        
#(2) Generating a WordCloud image
wordcloud = WordCloud(max_font_size = 80, collocations = False).generate(contents) # Setting lower max_font_size to 80

#(3) Display the generated image:
%matplotlib inline
plt.figure(figsize = (15, 10))
plt.imshow(wordcloud)
plt.axis("off")
plt.savefig('WordCloud2.pdf')
plt.show()
```
